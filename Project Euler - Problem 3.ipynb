```python
import math

count = 3
primes=[]
while True:
        isprime = True
        
        for x in range(2, int(math.sqrt(count) + 1)):
            if count % x == 0: 
                isprime = False
                break
        
            if isprime:
              primes.append(count)
        
        count += 1
        
        if count > math.sqrt(600851475143):
            break
            
       

  

        
```
